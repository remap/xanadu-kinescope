# Cleaning Skeletons with Center Shifting

## 1. Shift Skeletons and Flatten Arrays

```python
import numpy as np

def align_and_scale_skeletons_flat(flat_kp1, flat_kp2, spine_index=1):
    """
    Align and scale two skeletons based on a central keypoint (e.g., spine) for flattened keypoint arrays.

    Parameters:
        flat_kp1 (np.ndarray): First skeleton's keypoints in flat COCO 17 format.
        flat_kp2 (np.ndarray): Second skeleton's keypoints in flat COCO 17 format.
        spine_index (int): Index of the central spine keypoint (default: 1, as per COCO).

    Returns:
        scaled_flat_kp2 (np.ndarray): Keypoints of the second skeleton after alignment and scaling, in flat format.
    """
    def get_bounding_box(flat_kp):
        """Compute the bounding box of visible keypoints from flat array."""
        x_coords, y_coords = [], []
        for i in range(18):  # COCO 17 keypoints
            x, y, v = flat_kp[i * 3:(i * 3) + 3]
            if v > 0:  # Visible keypoints
                x_coords.append(x)
                y_coords.append(y)
        if not x_coords or not y_coords:
            raise ValueError("No visible keypoints found.")
        return (min(x_coords), min(y_coords), max(x_coords), max(y_coords))

    def scale_keypoints(flat_kp, scale_w, scale_h, center):
        """Scale keypoints in flat array format around a central point."""
        scaled = flat_kp.copy()
        for i in range(17):
            x, y, v = scaled[i * 3:(i * 3) + 3]
            if v > 0:  # Only scale visible keypoints
                scaled[i * 3] = center[0] + (x - center[0]) * scale_w
                scaled[i * 3 + 1] = center[1] + (y - center[1]) * scale_h
        return scaled

    # Extract spine keypoints for alignment
    center1 = flat_kp1[spine_index * 3:(spine_index * 3) + 2]
    center2 = flat_kp2[spine_index * 3:(spine_index * 3) + 2]

    # Align skeletons by translating keypoints2 to match keypoints1
    aligned_flat_kp2 = flat_kp2.copy()
    for i in range(18):
        x, y, v = aligned_flat_kp2[i * 3:(i * 3) + 3]
        if v > 0:  # Only translate visible keypoints
            aligned_flat_kp2[i * 3] += (center1[0] - center2[0])
            aligned_flat_kp2[i * 3 + 1] += (center1[1] - center2[1])

    # Calculate bounding boxes
    bbox1 = get_bounding_box(flat_kp1)
    bbox2 = get_bounding_box(aligned_flat_kp2)

    # Calculate scaling factors for width and height
    width1, height1 = bbox1[2] - bbox1[0], bbox1[3] - bbox1[1]
    width2, height2 = bbox2[2] - bbox2[0], bbox2[3] - bbox2[1]
    scale_w, scale_h = width1 / width2, height1 / height2

    # Scale keypoints2 to match keypoints1
    scaled_flat_kp2 = scale_keypoints(aligned_flat_kp2, scale_w, scale_h, center1)

    return scaled_flat_kp2

# Example usage:
# Skeleton keypoints in flat COCO 17 format [x0, y0, v0, ..., x16, y16, v16] (replace these with actual data)
# flat_kp1 = np.array([
#     100, 200, 1, 110, 220, 1, 120, 240, 1, 130, 260, 1, 140, 280, 1, 150, 300, 1,
#     160, 320, 1, 170, 340, 1, 180, 360, 1, 190, 380, 1, 200, 400, 1, 210, 420, 1,
#     220, 440, 1, 230, 460, 1, 240, 480, 1, 250, 500, 1, 260, 520, 1
# ])

# flat_kp2 = np.array([
#     50, 100, 1, 55, 110, 1, 60, 120, 1, 65, 130, 1, 70, 140, 1, 75, 150, 1,
#     80, 160, 1, 85, 170, 1, 90, 180, 1, 95, 190, 1, 100, 200, 1, 105, 210, 1,
#     110, 220, 1, 115, 230, 1, 120, 240, 1, 125, 250, 1, 130, 260, 1
# ])

flat_kp1 = np.array(

  [763.84375, 285.37832642, 2, 769.61334229, 319.89837646, 2, 739.19055176, 315.79663086, 2, 729.74664307, 353.73120117, 2, 720.88989258, 389.87475586, 2, 800.57299805, 324.07250977, 2, 802.06512451, 363.5302124, 2, 804.22729492, 401.43383789, 2, 738.26086426, 410.6293335, 2, 734.14788818, 496.0484314, 2, 729.84399414, 558.59570312, 2, 774.0682373, 416.28106689, 2, 771.40209961, 501.16693115, 2, 768.03521729, 569.67388916, 2, 760.1305542, 280.47171021, 2, 771.69519043, 280.60391235, 2, 753.53033447, 287.54129028, 2, 785.41778564, 288.09906006, 2],

)

flat_kp2 = np.array(
    [399.12997437, 305.73544312, 2, 392.01922607 ,329.26657104, 2, 365.07073975, 326.15209961, 2, 362.32302856, 358.77740479, 2, 360.36807251, 388.66491699, 2, 417.88824463, 332.2562561, 2, 420.84381104, 363.815979, 2, 422.53997803, 359.37200928, 2, 366.60964966, 411.51760864, 2, 365.30502319, 473.81097412, 2, 364.19903564, 526.12854004, 2, 399.89993286, 413.20977783, 2, 395.24258423, 474.9335022, 2, 391.13439941, 516.6038208, 2, 394.48440552, 300.61260986, 2, 402.50976562, 301.73754883, 2, 384.09439087, 302.30859375, 2, 406.32385254, 305.35794067, 2],

)

# Align and scale flat_kp2 to flat_kp1
scaled_flat_kp2 = align_and_scale_skeletons_flat(flat_kp1, flat_kp2)
print(scaled_flat_kp2)
flat_kp2 = scaled_flat_kp2
```

## 3. Calculate IOU and Ensure it is > Than COCO's 0.5 Threshold

```python
def compute_iou(box1, box2):
    x1, y1, w1, h1 = box1
    x2, y2, w2, h2 = box2

    # Compute coordinates of intersection rectangle
    xi1 = max(x1, x2)
    yi1 = max(y1, y2)
    xi2 = min(x1 + w1, x2 + w2)
    yi2 = min(y1 + h1, y2 + h2)

    # Compute area of intersection
    inter_width = max(0, xi2 - xi1)
    inter_height = max(0, yi2 - yi1)
    inter_area = inter_width * inter_height

    # Compute area of both boxes
    box1_area = w1 * h1
    box2_area = w2 * h2

    # Compute IoU
    union_area = box1_area + box2_area - inter_area
    iou = inter_area / union_area if union_area > 0 else 0
    return iou

bbox_gt = [727.1870261477947, 283.1524667489191, 83.33740233999993, 289.2021789500001]
bbox_dt = [720.88989258, 280.47171021, 83.33740233999993, 289.20217895]
print("IoU:", compute_iou(bbox_gt, bbox_dt))
```
